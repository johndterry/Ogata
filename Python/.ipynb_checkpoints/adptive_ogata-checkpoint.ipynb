## integrators

```python
def Quadinversion(w, q, nu, tau):
    quadreturn = quad(lambda bT: jv(nu,q*bT)*w(bT),0,np.inf, epsabs = 0, epsrel = tau)
    return 1/(2*np.pi)*quadreturn[0], 1/(2*np.pi)*quadreturn[1]
```

```python
def transform(f, p):
    return f(np.tan(p))*(1/np.cos(p))**2

def MCinversion(f, q, nu, m):
    integ = vegas.Integrator([[0, np.pi/2.0]])
    result = integ(lambda p: 1/(2*np.pi)*transform(f, p)*jv(nu, q*np.tan(p)), nitn=10, neval=int(m))[0]
    lst = str(result).replace('+-', '(').replace(')', '(').split('(')
    num_zeros = 0
    if str(lst[0][0]) == '-':
        num_zeros = -1
    if len(lst[1]) == 3:
        integral = float(lst[0])
        error = float(lst[1])
    else:
        num_zeros = num_zeros+len(lst[0])-4
        lst[1] = '0.'+lst[1].zfill(num_zeros+2)
        integral = float(lst[0])
        error = float(lst[1])
    return integral, error
```

## toy TMD

```python
def Wtilde(bT,Q,sigma):
    global counter
    counter += 1
    M=1.0/Q
    V=1/sigma**2
    b=0.5*(-M+np.sqrt(M**2+4*V))
    a=V/b**2
    return bT**(a-1)*np.exp(-bT/b)/b**a/spec.gamma(a)

def W(qT, Q, sigma, nu):
    M=1.0/Q
    V=1/sigma**2
    b=0.5*(-M+np.sqrt(M**2+4*V))
    a=V/b**2
    return 1/(2*np.pi)*spec.gamma(a+nu)/spec.gamma(a)*(b*qT/2.0)**nu*spec.hyp2f1((a+nu)/2.0, (a+nu+1.0)/2.0, nu+1.0, -qT**2.0*b**2.0)/spec.gamma(nu+1.0)

def Wtildenew(bT,Q,sigma):
    global counter
    counter += 1
    return b*np.exp(-b**2/(4*sigma**2))

def Wtildenew(bT,Q,sigma):
    global counter
    counter += 1
    sigma2=0.57+0.12/(0.15)**2
    return b*np.exp(-b**2/(4*sigma2))
```

# data generator

```python
def gen_data(nu,Q,q,sigma,epss,ns_vegas,ns_ogata):
    global counter
    exact=W(q, Q, sigma, nu)
    w = np.vectorize(lambda bT: Wtilde(bT, Q, sigma))

    data={}
    
    data['quad']={'N':[],'ratio':[],'error':[]}
    for eps in epss:
        counter = 0
        result=Quadinversion(w, q, nu, eps)
        data['quad']['N'].append(counter/2)
        data['quad']['ratio'].append(result[0]/exact)
        data['quad']['error'].append(result[1]/exact)

    data['vegas']={'N':[],'ratio':[],'error':[]}
    for n in ns_vegas:
        counter = 0
        result=MCinversion(w, q, nu, int(n)/10)
        data['vegas']['N'].append(counter/2)
        data['vegas']['ratio'].append(result[0]/exact)
        data['vegas']['error'].append(result[1]/exact)
        
    data['ogata']={'N':[],'ratio':[],'error':[]}
    for n in ns_ogata:
        counter = 0
        result=adog(w, q, nu, int(int(n)/10), Q**2)
        data['ogata']['N'].append(int(counter/6))
        data['ogata']['ratio'].append(result[0]/exact)
        data['ogata']['error'].append(result[1]/exact)
    return data
```

```python
nu = 0
Q = 1.0
q = 1e-5
sigma = 2.5
epss=10**np.linspace(-2,0,100)[::-1]
ns_vegas=np.arange(10,500,10)
ns_ogata=np.arange(40,1800,10)
data0=gen_data(nu,Q,q,sigma,epss,ns_vegas,ns_ogata)
```

# Old Algorithm

```python
def ogata(f,h,N, nu):
    zeros=jn_zeros(nu,N)
    xi=zeros/np.pi
    Jp1=jv(nu+1,np.pi*xi)
    w=yv(nu, np.pi * xi) / Jp1
    get_psi=lambda t: t*np.tanh(np.pi/2*np.sinh(t))
    get_psip=lambda t:np.pi*t*(-np.tanh(np.pi*np.sinh(t)/2)**2 + 1)*np.cosh(t)/2 + np.tanh(np.pi*np.sinh(t)/2)
    knots=np.pi/h*get_psi(h*xi)
    Jnu=jv(nu,knots)
    psip=get_psip(h*xi)
    F=f(knots)
    return np.pi*np.sum(w*F*Jnu*psip)

def Ogatainversion(w, xmin, xmax, qT, nu):
    h,N=get_ogata_params(w, xmin, xmax, qT, nu)
    return 1/(2*np.pi)*ogata(lambda x: w(x/qT)/qT,h,N, nu), h, N

def find_bmax(f, bmin, bmax, bnum = 20):
    bT = np.exp(np.linspace(np.log(bmin), np.log(bmax), bnum))
    f_bT = [f(b) for b in bT]
    imax = [i for i, j in enumerate(f_bT) if j == max(f_bT)][0]
    b_list_max = bT[imax]
    bmax = fmin(lambda x: -f(x), b_list_max, disp = False)
    return bmax[0]

def get_ogata_params_b(w, bmin, bmax, qT, nu):
    zero1 = jn_zeros(nu, 1)[0]
    h = fsolve(lambda h: bmin-zero1/qT*np.tanh(np.pi/2*np.sinh(h/np.pi*zero1)), bmin)[0]
    k = fsolve(lambda k: bmax-np.pi/qT*k*np.tanh(np.pi/2*np.sinh(h*k)), bmax)[0]
    if k<0:
        k = fsolve(lambda k: bmax-np.pi/qT*k*np.tanh(np.pi/2*np.sinh(h*k)), -k)[0]
    N = int(k+1)
    return h,N

def compare(w, ib, it, bc, qT, nu, exact, lib):
    ccut = 2.0
    #Original b range
    bmin = bc/((ccut)**ib)
    bmax = bc*ccut**it
    h,N=get_ogata_params_b(w, bmin, bmax, qT, nu)
    #Double bmax
    bmin1 = bc/((ccut)**ib)
    bmax1 = bc*ccut**it*ccut
    h1,N1=get_ogata_params_b(w, bmin1, bmax1, qT, nu)
    #Half bmin
    bmin2 = bc/((ccut)**ib)/ccut
    bmax2 = bc*ccut**it
    h2,N2=get_ogata_params_b(w, bmin2, bmax2, qT, nu)
    try:
        val1 = lib[str(bmin1)+','+str(bmax1)]
    except:
        val1 = 1/(2*np.pi)*ogata(lambda x: w(x/qT)/qT,h1,N1, nu)
        lib[str(bmin1)+','+str(bmax1)] = val1
    try:
        val2 = lib[str(bmin2)+','+str(bmax2)]
    except:
        val2 = 1/(2*np.pi)*ogata(lambda x: w(x/qT)/qT,h2,N2, nu)
        lib[str(bmin2)+','+str(bmax2)] = val2
    return abs((val1-exact)/exact)<abs((val2-exact)/exact), h, N, h1, N1, h2, N2, lib

def AdaptOgatainversion(w, qT, nu, imax, Nmax, exact, ib = 1, it = 1, bpeak = [], lib = {}):
    if ib+it< imax:
        _lib = lib
        if not bpeak:
            bc = find_bmax(w, 1e-2, 1e1)
        else:
            bc = bpeak[0]
        _bool, h, N, h1, N1, h2, N2, __lib = compare(w, ib, it, bc, qT, nu, exact, _lib)
        if N2<Nmax:
            if _bool == True:
                return AdaptOgatainversion(w, qT, nu, imax, Nmax, exact, ib = ib+0, it = it+1, bpeak = [bc], lib = __lib)
            else:
                return AdaptOgatainversion(w, qT, nu, imax, Nmax, exact, ib = ib+1, it = it+0, bpeak = [bc], lib = __lib)
        else:
            return 1/(2*np.pi)*ogata(lambda x: w(x/qT)/qT,h,N, nu), N
    else:
        print('imax reached')

def get_ogata_params(w, xmin, xmax, qT, nu):
    zero1 = jn_zeros(nu, 1)[0]
    h = fsolve(lambda h: xmin-zero1*np.tanh(np.pi/2*np.sinh(h/np.pi*zero1)), xmin)[0]
    k = fsolve(lambda k: xmax-np.pi*k*np.tanh(np.pi/2*np.sinh(h*k)), xmax)[0]
    N = int(k)
    return h,N
```
